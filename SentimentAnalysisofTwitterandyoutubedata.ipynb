# 2. Run VADER over evaluation texts

## 2.1 Run VADER over the first tweet

In this task you should use VADER for sentiment analysis. For this we use the `vaderSentiment` package. You first have to intatiate a new `SentimentIntensityAnalyzer` and use the `polarity_scores` method of it for the analysis. Apply this for the first tweet. Is it a good classification?

[Here](https://github.com/cjhutto/vaderSentiment) under 'Code Examples' you can find some example code how to use this package.

```python
# Your Code goes here!
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
sentences = ["?RT @justinbiebcr: The bigger the better....if you know what I mean ;)"]
#Initializing
sidobj = SentimentIntensityAnalyzer()




for sentence in sentences:
    vs = sidobj.polarity_scores(sentence)
    print("{:-<65} {}".format(sentence, str(vs)))


'''
The Sentiment of the tweet is being classified as {'neg': 0.0, 'neu': 0.853, 'pos': 0.147, 'compound': 0.2263}. 
The compound of 0.226 shows that the sentiment of the comment is slightly positive as it is above 0.05.
'''




```

## 2.2 Run VADER over each text

Now use VADER for all the text data of the Twitter and the Youtube dataframe. Create a new column in the dataframes called `VADER_compound` where you save the `compound` result (look at the output dictonary of the `polarity_scores` method).

*Important: Make sure `compound` is a float*



```python
# Your Code goes here!
#Initializing
sidobj = SentimentIntensityAnalyzer()

youtube = pd.DataFrame(youtubecomments)
youtube ['VADER_compound'] = youtube['text'].apply(
    lambda x:sidobj.polarity_scores(str(x))["compound"]
)

youtube
```

```python

twitter = pd.DataFrame(twittersentiment)
twitter['VADER_compound'] = twitter['text'].apply(
   lambda x:sidobj.polarity_scores(str(x))["compound"]
)

twitter

```

## 2.3 VADER as a classifier

To get the three Classes `Positive`, `Negative` and `Neutral` we use the compound score with the following thresholds:

* `compound > 0.5`: `"Positive"`
* `compound < -0.5`: `"Negative"`
* `else`: `"Neutral"`

Create, for both datasets, a new column called `VADER_class` which contains the three computed classes.

```python
# Your Code goes here!

youtube ['VADER_class'] = youtube["VADER_compound"].apply(
    lambda x: "positive" if x >= 0.5 else "negative" if x <= -0.5 else "neutral"
)

youtube
```

```python

twitter ['VADER_class'] = twitter["VADER_compound"].apply(
    lambda x: "positive" if x >= 0.5 else "negative" if x <= -0.5 else "neutral"
)

twitter
```

## 4.2 Evaluate on Twitter
For all evaluations, use the `label` column as the true value.  
If you want to know more information on the metrics, you can read [this website](https://developers.google.com/machine-learning/crash-course/classification/accuracy-precision-recall).  
First, let's calculate the accuracy for all three classifiers (vader, bert, google) on the Twitter dataset.

```python
# Your Code goes here!
from sklearn.metrics import accuracy_score


twitter ['VADER_class'] = twitter["VADER_compound"].apply(
    lambda x: "positive" if x >= 0.5 else "negative" if x <= -0.5 else "neutral"
)

twitter ['BERT_class'] = twitter['text'].astype(str).apply(
    lambda b: sentimentpipeline (b)[0]["label"]
)

twitter ['GoogleNPL_class'] = twitter["googleScore"].apply(
    lambda x: "positive" if x >= 0.3 else "negative" if x <= -0.3 else "neutral"
)

twitter
```

```python
twitter["VADER_class"] = twitter["VADER_class"].str.capitalize()
twitter["BERT_class"] = twitter["BERT_class"].str.capitalize()
twitter["GoogleNPL_class"] = twitter["GoogleNPL_class"].str.capitalize()


accuracyvader= accuracy_score(twitter['label'], twitter['VADER_class'])
accuracybert= accuracy_score(twitter['label'], twitter['BERT_class'])
accuracygoogle= accuracy_score(twitter['label'], twitter['GoogleNPL_class'])

print (f"Accuracy for BERT:", accuracybert , "Accuracy for Vader: ", accuracyvader, "Accuracy for Google: ", accuracygoogle)
```

```python
#vader
ytrue = twitter['label']
ypred = twitter['VADER_class'].str.capitalize()

TP = ((ypred == "Positive" ) & (ytrue == "Positive")).sum()
FP = ((ypred == "Positive") & (ytrue != "Positive")).sum()
FN = ((ypred!= "Positive") & (ytrue == "Positive")).sum()


TP,FP,FN
```

```python
#bert
ytrue2 = twitter['label']
ypred2 = twitter['BERT_class'].str.capitalize()

TP2 = ((ypred2 == "Positive" ) & (ytrue2 == "Positive")).sum()
FP2 = ((ypred2 == "Positive") & (ytrue2 != "Positive")).sum()
FN2 = ((ypred2 != "Positive") & (ytrue2 == "Positive")).sum()


TP2,FP2,FN2
```

```python
#Google
ytrue3 = twitter['label']
ypred3 = twitter['GoogleNPL_class'].str.capitalize()

TP3 = ((ypred3 == "Positive" ) & (ytrue3 == "Positive")).sum()
FP3 = ((ypred3 == "Positive") & (ytrue3 != "Positive")).sum()
FN3 = ((ypred3 != "Positive") & (ytrue3 == "Positive")).sum()

TP3,FP3,FN3

```

Next calculate the precision of the `"Positive"` class for the Twitter data.  
This is calculated as follows:
$
\begin{align}
    precision = \frac{TP}{TP + FP}
\end{align}
$
*Note: Here the Positive samples are the one with the the class `"Positive"`*

```python
# Your Code goes here!

Precisionvader = TP/(TP + FP)
print("Precision for VADER", Precisionvader)

Precisionbert = TP2/(TP2 + FP2)
print("Precision for BERT", Precisionbert)

Precisiongoogle = TP3/(TP3 + FP3)
print("Precision for Google",Precisiongoogle)

```

Now calculate the recall score. This is done by:
$
\begin{align}
    recall = \frac{TP}{TP + FN}
\end{align}
$
*Note: Here the Positive samples are the one with the the class `"Positive"`*

```python
# Your Code goes here!
Recallvader = TP/(TP + FN)
print("Recall for VADER", Recallvader)

Recallbert= TP2/(TP2 + FN2)
print("Recall for BERT", Recallbert)

Recallgoogle = TP3/(TP3 + FN3)
print("Recall for Google", Recallgoogle)

```

Calculate the Recall and the Precision score now also for the negative class. The Precision is calculated as:
$

$
\begin{align}
    recall = \frac{TP}{TP + FN}

```python
# Your Code goes here!

#vader
ytrue = twitter['label']
ypred = twitter['VADER_class'].str.capitalize()

TP = ((ypred == "Negative" ) & (ytrue == "Negative")).sum()
FP = ((ypred == "Negative") & (ytrue != "Negative")).sum()
FN = ((ypred!= "Negative") & (ytrue == "Negative")).sum()


TP,FP,FN

Precisionvader = TP/(TP + FP)
print("Precision Negative for VADER", Precisionvader)

Recallvader = TP/(TP + FN)
print("Recall Negative for VADER", Recallvader)

```

Precision and Recall are both metrics for measuring the quality of classification models. 
Whereas precision measures the proportion of true positives (correct predictions) in relation to the total number of true (true positive and false positive) predictions, recall measures the amount of true positives in relation to the actual number of truths (true positives and false negatives). 

In this case precision would measure whether the model predicted e.g. the positive tweets correctly in relation to all positive predictions made by the model whereas recall would measure the positive tweets in relation to the total number of positive tweets in the dataset.
